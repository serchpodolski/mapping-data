```python
import pandas as pd
```

```python
wine_df = pd.read_csv('/workspaces/mapping-data/sample_data/wine-ratings.csv')
```

```python
wine_df.head(10)
```

```python
def wine_type(variety):
    if variety=='Red Wine':
        return 'Red'
    else:
        return 'White'
```

```python
wine_df['Type'] = wine_df['variety'].apply(wine_type)
wine_df['Valuable'] = wine_df['rating'].apply(lambda x: 'Pass' if x>90 and x<93 else 'Fail')
```

```python
wine_df.head(15)
```

```python
whites_df = wine_df[(wine_df['Type']=='White') & (wine_df['Valuable']=='Pass')]
```

```python
whites_df.head()
```

```python
import json
```

```python
whites_to_json = whites_df.to_json(orient='columns')
parsed_whites = json.loads(whites_to_json)
```

```python
with open('good_whites.json', 'w') as f:
    json.dump(parsed_whites, f, indent=4)
```
